```python
# importing two countries rounds2 and companies
import numpy as np
import pandas as pd
rounds2=pd.read_csv(r"E:\Case_Study\rounds2.csv",sep=",", encoding = "ISO-8859-1")
companies=pd.read_csv(r"E:\Case_Study\companies.txt",sep="\t", encoding = "ISO-8859-1")
pd.options.display.float_format = '{:,.1f}'.format
```

```python
#by splitting the company_permalink column to seperate the company field
rounds2['company']=rounds2['company_permalink'].str.split('/').apply(lambda x : x[2] if len(x) > 2 else x[0])
rounds2.head(2)
```

```python
#by splitting the permalink column to seperate the company field
companies['company'] = companies['permalink'].str.split('/').apply(lambda x : x[2] if len(x) > 2 else x[0])
companies.head(2)
```

```python
#encoding converstion and creating the master_frame
rounds2['company']=rounds2['company'].str.encode('utf-8').str.decode('ascii', 'ignore')
companies['company']=companies['company'].str.encode('utf-8').str.decode('ascii', 'ignore')

companies['company']=companies['company'].str.lower()
rounds2['company']=rounds2['company'].str.lower()

master_frame=pd.merge(rounds2,companies,how='inner')
master_frame
```

```python
# doing encoding converstion column level also..to ignore special characters
master_frame['company_permalink']=master_frame['company_permalink'].str.encode('utf-8').str.decode('ascii', 'ignore')
master_frame['permalink']=master_frame['permalink'].str.encode('utf-8').str.decode('ascii', 'ignore')
master_frame
```

```python
master_frame.shape
```

```python
# finding the null percentage for all columns
round(100*(master_frame.isnull().sum()/len(master_frame.index)),2)
```

```python
# As null percentage is in huge in raised_amount_usd column removing the records from master_frame part of the analysis
master_frame=master_frame[~master_frame['raised_amount_usd'].isnull()]
```

```python
#finding the primary_category_list by splitting the category_list field
master_frame['category_list'] = master_frame['category_list'].astype('str')
master_frame['primary_category_list'] = master_frame['category_list'].apply(lambda x : x.split('|')[0])
master_frame
```

```python
# importing the mapping file and finding the respective caterogy list
mapping=pd.read_csv(r"E:\Case_Study\original_files\mapping.csv",sep=",")
mapping.head()

mapping['main_sector']=np.nan
mapping.loc[mapping['Automotive & Sports']==1,'main_sector']='Automotive & Sports'
mapping.loc[mapping['Cleantech / Semiconductors']==1,'main_sector']='Cleantech / Semiconductors'
mapping.loc[mapping['Entertainment']==1,'main_sector']='Entertainment'
mapping.loc[mapping['Health']==1,'main_sector']='Health'
mapping.loc[mapping['Manufacturing']==1,'main_sector']='Manufacturing'
mapping.loc[mapping['News, Search and Messaging']==1,'main_sector']='News, Search and Messaging'
mapping.loc[mapping['Others']==1,'main_sector']='Others'
mapping.loc[mapping['Social, Finance, Analytics, Advertising']==1,'main_sector']='Social, Finance, Analytics, Advertising'
mapping[['category_list','main_sector']]
```

```python
# removing unncessary columns before merge into master_frame
mapping.drop(['Automotive & Sports','Cleantech / Semiconductors','Entertainment','Health','Manufacturing','Others',
             'Social, Finance, Analytics, Advertising', 'News, Search and Messaging', 'Blanks'], 
              axis=1,inplace=True)
```

```python
# joining the master_frame and mapping by category_list
master_frame =pd.merge(master_frame,mapping,how='inner',on='category_list')
master_frame
```
